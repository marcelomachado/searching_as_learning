```python
df = pd.read_csv('data/sal.csv', sep='\t')
df = df.drop(["execution_time", "action", "client_version", "course_id"], axis=1)
df.head()
```

```python
userids = list(df.user_id.unique())
```

```python
def get_search_sessions(df, userid):
    usersearchsessions = list(df[(df.user_id == userid) & (df.description == "searching")].session_id.unique())
    dss = df[(df.user_id == userid) & (df.session_id.isin(usersearchsessions))]
    
    return dss    
```
